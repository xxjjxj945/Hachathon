```python
def longitude(address):
  link_main = 'https://api-adresse.data.gouv.fr/search/?q='
  link_end = '&format=json&limit=1'
  link= link_main + address.replace(" ", "+")
  r = requests.get(link+link_end).json()
  longi=r['features'][0]['geometry']['coordinates'][-1]
  
  return longi
```

```python
def latitude(address):
  link_main = 'https://api-adresse.data.gouv.fr/search/?q='
  link_end = '&format=json&limit=1'
  link= link_main + address.replace(" ", "+")
  r = requests.get(link+link_end).json()
  lat=r['features'][0]['geometry']['coordinates'][0]
  
  return lat
```

```python
def coordonnées(adresse):
  link_main = "https://api-adresse.data.gouv.fr/search/?q="
  link = link_main + adresse.replace(" ", "+")
  r = requests.get(link).json()
  return r['features'][0]['geometry']['coordinates'][::-1]
```
